```python
import xarray as xr
import numpy as np
import xesmf as xe
```

```python
file = '/groups/ESS3/qtong/data/nemo2017/emis_mole_all_2017ann_US01_cmaq_cb6ae7_2017gb_17j_nosoaalk.ncf'
ds = xr.open_dataset(file)
```

```python
ds
```

```python
coords = '/scratch/averstyn/GRID_LATLON_CENTRAL_US01.ncf'
coords_ds = xr.open_dataset(coords)
coords_ds
```

```python
pm25 = ds['PM25ANN'][0,0,:,:]
lats = coords_ds['LAT'][0,0,:,:]
lons = coords_ds['LON'][0,0,:,:]
```

```python
new_ds = xr.Dataset(data_vars={'PM25':pm25},coords={'lat':lats,'lon':lons})
new_ds
```

```python
new_lat=np.arange(-90,90.5,0.5)
new_lon=np.arange(0,360.1,0.1)

ds_out = xr.Dataset({'lat': (['lat'], new_lat),
                    'lon': (['lon'], new_lon)
                    })
ds_out
```

```python
var = 'PM25'
regridder = xe.Regridder(new_ds,ds_out,'bilinear')
dr_out = regridder(new_ds[var]).to_dataset(name=var)
dr_out.reset_coords()
dr_out
```

```python
dr_out.to_netcdf('/scratch/averstyn/NEMO_latlon.nc')
```
